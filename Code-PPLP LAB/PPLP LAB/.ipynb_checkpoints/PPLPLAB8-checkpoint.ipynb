```python
data = pd.read_csv("diamonds.csv")
```

```python
price_val = data['price']
price_val
```

```python
normalized_price = (price_val - price_val.min())/(price_val.max()-price_val.min())
data['normalized_price']=normalized_price
data.sort_values(by='price')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

data = pd.read_csv('diamonds.csv')

plt.hist(data['cut'],color='darkblue',orientation='vertical')
plt.title('cut histogram')
plt.xlabel('cut')
plt.ylabel('frequency')
cut = ('ideal','Premium','Good','Very good','fair')
index = np.arange(len(cut))
plt.xticks(index,cut,rotation=0)
plt.show()
```

```python
print(data.corr())
plt.show()
```

```python
plt.matshow(data.corr())
plt.xticks(range(len(data.columns)), data.columns)
plt.yticks(range(len(data.columns)), data.columns)
plt.colorbar()
plt.show()
```

```python
sns.heatmap(data.corr())
```

```python
correlation_matrix = data.corr().round(2)
sns.heatmap(data=correlation_matrix,annot=True)
```
